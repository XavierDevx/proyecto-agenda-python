```python
class Contacto:
  # Constructor
  def __init__ (self, nombre, telefono, email, direccion):
    # Atributos
    self.nombre = nombre
    self.telefono = telefono
    self.email = email
    self.direccion = direccion

  def mostrar_contacto (self):
    print ("Nombre:", self.nombre)
    print ("Teléfono:", self.telefono)
    print ("Email:", self.email)
    print ("Dirección:", self.direccion)
```

Gestor de contactos:

```python
class GestionContactos:
  def __init__(self):
    self.contactos = []

  def agregar_contacto(self, contacto):
    self.contactos.append(contacto)

  def lista_contacto(self):
    for contacto_obj in self.contactos:
      contacto_obj.mostrar_contacto()

  def buscar_contacto(self, nombre_buscado):
    print(f"\n Buscando a {nombre_buscado}...")
    encontrado = False

    for contacto_obj in self.contactos:
      if contacto_obj.nombre.lower() == nombre_buscado.lower():
        contacto_obj.mostrar_contacto()
        encontrado = True

    if not encontrado:
      print(f"No se encontró a {nombre_buscado}.")

  def eliminar_contacto(self, nombre_buscado):
    print(f"\n Eliminando a {nombre_buscado}...")
    encontrado = False # Initialize 'encontrado'

    for contacto_obj in self.contactos:
      if contacto_obj.nombre.lower() == nombre_buscado.lower():
        self.contactos.remove(contacto_obj) # Actual removal
        print(f"{nombre_buscado} ha sido eliminado.")
        encontrado = True
        return # Exit after finding and removing the contact

    if not encontrado:
      print(f"No se encontró a {nombre_buscado}.")

  def editar_contacto(self, nombre_buscado, nuevo_nombre, nuevo_telefono, nuevo_email, nueva_direccion):
    print(f"\n Editando a {nombre_buscado}...")

    for contacto_obj in self.contactos:
      if contacto_obj.nombre.lower() == nombre_buscado.lower():
        contacto_obj.nombre = nuevo_nombre
        contacto_obj.telefono = nuevo_telefono
        contacto_obj.email = nuevo_email
        contacto_obj.direccion = nueva_direccion
        print(f"{nombre_buscado} ha sido editado.")
        return

    print(f"No se encontró a {nombre_buscado}.")
```

Prueba de concepto: Menú

```python
def inicio_interfaz():
  Mi_agenda = GestionContactos()

  while True:
    print("\nMenú de Agenda:")
    print("1. Agregar Contacto")
    print("2. Lista de Contactos")
    print("3. Buscar Contacto")
    print("4. Eliminar Contacto")
    print("5. Editar Contacto")
    print("6. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
      nombre = input("Nombre: ")
      telefono = input("Teléfono: ")
      email = input("Email: ")
      direccion = input("Dirección: ")
      nuevo_contacto = Contacto(nombre, telefono, email, direccion)
      Mi_agenda.agregar_contacto(nuevo_contacto)

    if opcion == "2":
      Mi_agenda.lista_contacto()
      for contacto_obj in Mi_agenda.contactos:
        contacto_obj.mostrar_contacto()

    if opcion == "3":
      nombre_buscado = input("Nombre del contacto a buscar: ")
      Mi_agenda.buscar_contacto(nombre_buscado)

    if opcion == "4":
      nombre_buscado = input("Nombre del contacto a eliminar: ")
      Mi_agenda.eliminar_contacto(nombre_buscado)

    if opcion == "5":
      nombre_buscado = input("Nombre del contacto a editar: ")
      nuevo_nombre = input("Nuevo nombre: ")
      nuevo_telefono = input("Nuevo teléfono: ")
      nuevo_email = input("Nuevo email: ")
      nueva_direccion = input("Nueva dirección: ")
      Mi_agenda.editar_contacto(nombre_buscado, nuevo_nombre, nuevo_telefono, nuevo_email, nueva_direccion)

    if opcion == "6":
      print("Saliendo de la agenda. ¡Hasta luego!")
      break

if __name__ == "__main__":
    inicio_interfaz()
```
